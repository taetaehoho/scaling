```python
import pandas as pd
```

```python
df = pd.read_csv("op_bytes_per_batch.csv")
df['block_time'] = pd.to_datetime(df['block_time'])
df = df.set_index('block_time')
daily_counts = df.resample('D').size()

# 1. Convert daily_counts to a DataFrame
daily_counts_df = daily_counts.to_frame(name='count')

# 2. Calculate 30-day moving average and 30-day median
daily_counts_df['30d_moving_avg'] = daily_counts_df['count'].rolling(window=30).mean()
daily_counts_df['30d_moving_median'] = daily_counts_df['count'].rolling(window=30).median()

# 3. Calculate overall median, 25th percentile, 75th percentile, and average
overall_median = daily_counts_df['count'].median()
q25 = daily_counts_df['count'].quantile(0.25)
q75 = daily_counts_df['count'].quantile(0.75)
overall_average = daily_counts_df['count'].mean()

# Display the results
print(daily_counts_df.head())  # to see the first few rows with 30-day calculations
print(f"Overall Median: {overall_median}")
print(f"25th Percentile: {q25}")
print(f"75th Percentile: {q75}")
print(f"Overall Average: {overall_average}")
```
