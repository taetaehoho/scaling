```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Assume a range of x values (replace with your own range if necessary)
x_values = np.linspace(0, 100000, 100)  # For example, x values from 0 to 1000

# From OP and ARB linear regressions

# OP 
y_values1 = 190.38 * x_values + 1585859.20

# Arb
y_values2 = 124.32 * x_values + -20303366.81

# zksync regression data 
y_values3 = 215.14 * x_values + -5245902.85

# Create a new figure
plt.figure(figsize=(10, 6))

# Plot the first equation with a red line
sns.lineplot(x=x_values, y=y_values1, label='y = 5362.72x + 93570156.15', color='red')

# Plot the second equation with a blue line
sns.lineplot(x=x_values, y=y_values2, label='y = 1678.90x + -30482705.52', color='blue')

# Plot the second equation with a black line
sns.lineplot(x=x_values, y=y_values3, label='y = 0.23x - 21215.88', color='black')

# Set the labels and title
plt.xlabel('total_transactions')
plt.ylabel('l1_gas_total')
plt.title('OP vs ARB vs Zksync tx vs l1_gas')

# Display the plot
plt.legend()
plt.grid(True)
plt.show()
```
